## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
ages = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# create a copy of the orignal df to keep original intact
purchase_age_df = purchase_data.copy()

# Create bins for different age groups and adds them to the df
purchase_age_df['Age Ranges'] = pd.cut(purchase_age_df["Age"], bins, labels=ages)

# Create a groupby object to make perform calculations based on age groups.
# ? Not sure if I was supposed to use groupby, but it makes it easy/clean.
age_group = purchase_age_df.groupby('Age Ranges')

# Perform calculations on the groupby object
purchase_count = age_group['Item ID'].count()
purchase_avg = age_group['Price'].mean()
purchase_total = age_group['Price'].sum()
unique_people = age_group['SN'].nunique()

# Create the df with the calculations we performed above.
summary_df = pd.DataFrame({'Purchase Count': purchase_count,
                          'Average Purchase Price': purchase_avg,
                          'Total Purchase Value': purchase_total})
summary_df['Average Total per Person'] = purchase_total / unique_people

# Add formatting to columns
summary_df['Average Purchase Price'] = summary_df['Average Purchase Price'].map("${:.2f}".format)
summary_df['Total Purchase Value'] = summary_df['Total Purchase Value'].map("${:,.2f}".format)
summary_df['Average Total per Person'] = summary_df['Average Total per Person'].map("${:.2f}".format)
summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a smaller df with the three columns from the original df
reduced_df = purchase_data[['Item ID', 'Item Name', 'Price']]

# Create a groupby object with both 'Item ID' and 'Item Name'
reduced_group = reduced_df.groupby(['Item ID', 'Item Name'])

# Perform calculations on the groupby object
item_count = reduced_group['Item Name'].count()
item_price = reduced_group['Price'].mean()

# Create a df with the calculations from above
reduced_summary = pd.DataFrame({'Purchase Count': item_count,
                               'Item Price': item_price})

# Add the column Total Purchase Vale and calculate the value
reduced_summary['Total Purchase Value'] = (item_count * item_price)
# Add Formatting
reduced_summary['Item Price'] = reduced_summary['Item Price'].map('${:.2f}'.format)
reduced_summary['Total Purchase Value'] = reduced_summary['Total Purchase Value'].map('${:.2f}'.format)
# Sort the df by purchase count in descending order
reduced_summary.sort_values('Purchase Count', ascending=False, inplace=True)
reduced_summary.head()
```
